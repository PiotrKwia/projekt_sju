## 1. Klasyczna entropia Shannona  
Dla zmiennej losowej $X$ przyjmującej wartości $x_i$ z prawdopodobieństwami $p_X(i)$ entropia Shannona to  
$$
H(X)\;=\;- \sum_i p_X(i) \,\log p_X(i)\,.  
$$  
Ma ona interpretację średniej niepewności lub informacji potrzebnej do opisania wyniku jednego pobrania z rozkładu $p_X$.
Jednostką entropii Shannona jest zazwyczaj **bit** (dla logarytmu o podstawie 2), **nat** (dla logarytmu naturalnego) lub **hartley** (ew. **dit**) dla logarytmu o podstawie 10. 

## 2. Operator gęstości jako uogólnienie zmiennej losowej  
Stan kwantowy mieszaniny opisuje się operatorem gęstości  
$$
\rho \;=\;\sum_i p_i\,\bigl|\psi_i\bigr\rangle\bigl\langle\psi_i\bigr|
\,,\qquad
p_i\ge0,\;\sum_i p_i=1
$$  
ale wartości $p_i$ zależą od konkretnej dekompozycji.
Operator gęstości stanów jest operatorem hermitowskim o wartości śladu $\mathrm{Tr}=1$. Warto pamiętać, że wartość śladu nie zależy od wyboru bazy, bowiem
$$
\mathrm{Tr}\!\left(U \rho U^\dagger\right) = \mathrm{Tr}\!\left(\rho U^\dagger U\right) = \mathrm{Tr}\!\left(\rho\right) = 1
$$
Operator gęstości stanów, jak każdy operator hermitowski, ma dekompozycję spektralną postaci
$$
\rho \;=\;\sum_i \lambda_i\,\bigl|\phi_i\bigr\rangle\bigl\langle\phi_i\bigr|
\,,\qquad
\lambda_i\ge0,\;\sum_i\lambda_i=1
$$
gdzie $\ket{\phi_i}$ jego wektory własne i $\{\lambda_i\}$ - przypisane im wartości własne. Wektory własne tworzą bazę przestrzenie są bowiem ortogonalne i unormowane
$$
\left\langle\phi_k\bigl|\phi_l\right\rangle = \delta_{k,l}
$$
Wartości własne operatora hermitowskiego są rzeczywiste i nieujemne, a na mocy poprzedniej uwagi $\sum_i \lambda_i = 1$.
Oznacza to, że **wartości własne operatora gęstości stanów definiują pewien rozkład prawdopodobieństwa**.

## 3. Definicja entropii von Neumanna

Entropia von Neumanna to naturalne uogólnienie entropii Shannona, wykorzystujące wartości własne operatora gęstości i mierzące “niepewność” stanu kwantowego
$$
S(\rho) \;=\; -\,\mathrm{Tr}\!\bigl(\rho\,\log\rho\bigr)
\;=\;
-\,\sum_i \lambda_i\,\log\lambda_i
$$  
gdzie $\log\rho$ działa na wartościach własnych w dekompozycji spektralnej.

Entropia vin Neumanna można interpretować jako
1. **Miarę mieszania:**  
  - $S(\rho)=0$ $\;\iff\;$ $\rho$ jest czystym stanem.  
  - Maksimum $S(\rho)=\log d$ osiąga stan maksymalnie zmieszany $\rho=\tfrac1d I$.  
2. **Informacja o pomiarze:**  
  W pomiarze w bazie $\{|\phi_i\rangle\}$ wynik $i$ wystąpi z prawdopodobieństwem $\lambda_i$, a średnia informacja to $-\,\sum\lambda_i\log\lambda_i$.  

Wartość entropii von Neumanna **NIE** zależy od wyboru bazy
$$S(U\rho\,U^\dagger)=S(\rho).$$

## 4. Porównanie właściwości entropii klasycznej i kwantowej

To zestawienie ilustruje, że choć wiele struktur formalnych się pokrywa, to w świecie kwantowym pojawiają się zupełnie nowe zjawiska (ujemna entropia warunkowa w wyniku splątania stanów), które nie mają odpowiednika w klasycznej teorii prawdopodobieństwa.

- **Układ klasyczny**.  Mamy zmienne losowe $X,Y$ z rozkładem łącznym $p_{XY}(x,y),$ który definiuje rozkłady brzegowe $p_X(x)=\sum_y p_{XY}(x,y)$, $p_Y(y)=\sum_x p_{XY}(x,y)$.
- **Układ kwantowy**. Stan układu dwuskładnikowego opisuje operator gęstości stanów $\rho_{AB}$, a stany brzegowe otrzymujemy przez ślad częściowy po odrzuconym układzie: 
  $\rho_A=\mathrm{Tr}_B(\rho_{AB})$, $\rho_B=\mathrm{Tr}_A(\rho_{AB})$.

### Czystość i mieszanie

- **Klasycznie:**  
  - Rozkład jest czysty (deterministyczny) $\iff$ istnieje dokładnie jedna para $(x_0,y_0)$ z $p_{XY}(x_0,y_0)=1$.  
  - Wtedy $H(X,Y)=0$, a brzegowo również $H(X)=H(Y)=0$.

- **Kwantowo:**  
  - Stan czysty: $\rho_{AB}=\ket{\Psi_{AB}}\bra{\Psi_{AB}}$, wtedy $S(\rho_{AB})=0$.  
  - **Różnica kluczowa:** stan czysty dwuskładnikowego układu może dawać mieszane redukcje! tj. $S(\rho_{AB})=0$ ale $S(\rho_A)=S(\rho_B)>0$. Jest to efekt **splątania**, którego nie ma w klasycznej teorii.

Na przykład dla pary EPR 
$$\ket{\beta_{0,0}}_{AB}=CX_{A\to B} H_A \ket{0_A}\ket{0_B} = \frac{1}{\sqrt{2}}\left(\ket{0_A}\ket{0_B}+\ket{1_A}\ket{1_B}\right)$$

Niwch $\rho_{AB}\;=\;\ket{\beta_{0,0}}\bra{\beta_{0,0}}$. Pary EPR $\ket{\beta_{m,n}}$ tworzą bazę w przestrzeni dwukubitowej, zatem wyrażenie to jest jednocześnie rozkładem spektralnym z jedną wartością własną $\lambda_0=1$. W efekcie entropia von Neumanna $S(\rho_{AB})=0$.
W bazie obliczeniowej $\{\ket{0_A 0_B},\ket{0_A 1_B},\ket{1_A 0_B},\ket{1_A 1_B}\}$ mamy 
$$
\rho_{AB}
=\ket{\beta_{0,0}}_{AB}\bra{\beta_{0,0}}_{AB}
=\frac12\bigl(\ket{0_A 0_B}+\ket{1_A 1_B}\bigr)\bigl(\bra{0_A 0_B}+\bra{1_A 1_B}\bigr)
=\frac12\bigl(\ket{0_A 0_B}\bra{0_A 0_B}+\ket{0_A 0_B}\bra{1_A 1_B}+\ket{1_A 1_B}\bra{0_A 0_B}+\ket{1_A 1_B}\bra{1_A 1_B}\bigr).
$$

Ale 
$$
\rho_A
\;=\;
\mathrm{Tr}_B(\rho_{AB}) = \sum_k \bra{k_B}\rho_{AB}\ket{k_B} = $$
$$
= 
\frac{1}{2} \bra{0_B}\bigl(\ket{0_A 0_B}\bra{0_A 0_B}+\ket{0_A 0_B}\bra{1_A 1_B}+\ket{1_A 1_B}\bra{0_A 0_B}+\ket{1_A 1_B}\bra{1_A 1_B}\bigr)\ket{0_B} + \frac{1}{2} \bra{1_B}\bigl(\ket{0_A 0_B}\bra{0_A 0_B}+\ket{0_A 0_B}\bra{1_A 1_B}+\ket{1_A 1_B}\bra{0_A 0_B}+\ket{1_A 1_B}\bra{1_A 1_B}\bigr)\ket{1_B}
=
$$
$$
\;=\;
\frac{1}{2} \ket{0_A} \bra{0_B}\ket{0_B}\bra{0_A} \bra{0_B}\ket{0_B} + \frac{1}{2} \ket{1_A}\bra{1_B}\ket{1_B}\bra{1_A} \bra{1_B}\ket{1_B}
\;=\;
\frac{1}{2}\Bigl(\ket{0_A}\bra{A} + \ket{1_A}\bra{1_A}\Bigr).
$$
Zatem operator $\rho_A$ ma dwa wektory własne $\ket{0_A}$, $\ket{1_A}$ z przypisaną wartością własną $\lambda_0=\lambda_1=\frac12$.
W reprezentacji macierzowej mamy
$$
\rho_A = \begin{bmatrix} \frac12 & 0 \\ 0 & \frac12\end{bmatrix}
$$
Entropia von Neumanna tej macierzy to  
$$
S(\rho_A)
\;=\;
-\mathrm{Tr}\Bigl(\rho_A\;\log_2(\rho_A)\Bigr)
\;=\;
-\mathrm{Tr}\Bigl(\begin{bmatrix} \frac12 & 0 \\ 0 & \frac12\end{bmatrix}\;\log_2(\begin{bmatrix} \frac12 & 0 \\ 0 & \frac12\end{bmatrix})\Bigr)
\;=\;
-\mathrm{Tr}\Bigl(\begin{bmatrix} \frac12 & 0 \\ 0 & \frac12\end{bmatrix}\;\begin{bmatrix} \log_2\frac12 & 0 \\ 0 & \log_2 \frac12\end{bmatrix}\Bigr)
\;=\;
-\mathrm{Tr}\Bigl(\begin{bmatrix} \frac12 \log_2\frac12 & 0 \\ 0 & \frac12 \log_2 \frac12\end{bmatrix}\Bigr)
\;=\;
-\;2\cdot\tfrac12\;\log_2\tfrac12
\;=\;
1\;\text{bit}.
$$  

Analogicznie odrzucenie podsystemu B daje
$$
\rho_B
\;=\;
\mathrm{Tr}_A(\rho_{AB})
\;=\;
\begin{bmatrix} \frac12 & 0 \\ 0 & \frac12\end{bmatrix}
$$  
więc  
$$
S(\rho_B)
\;=\;
1\;\text{bit}.
$$  


## Entropie i ich własności

- **Entropia Shannona:**  
  $$H(X,Y)=-\sum_{x,y}p_{XY}(x,y)\,\log p_{XY}(x,y),$$  
  $$H(X)= -\sum_x p_X(x)\,\log p_X(x),\quad H(Y)= -\sum_y p_Y(y)\,\log p_Y(y).$$  
  Właściwości:  
  $$H(X,Y)\le H(X)+H(Y),$$  
  równość $\iff$ $X,Y$ są niezależne:  
  $$p_{XY}(x,y)=p_X(x)\,p_Y(y).$$

- **Entropia von Neumanna:**  
  $$S(\rho_{AB})=-\mathrm{Tr}\!\bigl[\rho_{AB}\log\rho_{AB}\bigr],$$  
  $$S(\rho_A)=-\mathrm{Tr}\!\bigl[\rho_A\log\rho_A\bigr],\quad S(\rho_B)=-\mathrm{Tr}\!\bigl[\rho_B\log\rho_B\bigr].$$  
  Podaddytywność:  
  $$S(\rho_{AB})\le S(\rho_A)+S(\rho_B),$$  
  równość $\iff$ $\rho_{AB}=\rho_A\otimes\rho_B$ (stan produktowy, czyli brak korelacji kwantowych).

### Wzajemna informacja

- **Klasyczna:** opisuje redukcję niepewności $X$ dzięki znajomości $Y$
  $$I(X;Y)=H(X)+H(Y)-H(X,Y)\;\ge0,$$  
  
- **Kwantowa:**  mierzy całkowite korelacje (klasyczne + kwantowe)
  $$I(A\!:\!B)=S(\rho_A)+S(\rho_B)-S(\rho_{AB})\;\ge0.$$  

### Entropia warunkowa

- **Klasyczna (zawsze $\ge0$):**  
  $$H(X|Y)=H(X,Y)-H(Y)\;\ge0.$$

- **Kwantowa (może być ujemna!):**  
  $$S(A|B)=S(\rho_{AB})-S(\rho_B).$$  
  Ujemna wartość $S(A|B)<0$ jest sygnałem splątania i ma zastosowanie np. w protokołach kwantowej teleportacji.

### Nierówność przetwarzania danych (DPI)

Nierównośc przetwarzania danych mówi, że na każdym etapie przetwarzania informacji możemy ją tracić (może wyciekać), ale nie może jej przybywać.

- **Klasycznie:** 
  Dla każdej mapy probabilistycznej (kanału klasycznego) $X\to X'$  
  $$I(X';Y)\le I(X;Y).$$

- **Kwantowo:**  
  Dla każdej mapy CPTP (kanału kwantowego) $\mathcal{E}_A$ działającej lokalnie na $A$:  
  $$I\bigl(\mathcal{E}_A(\rho_{AB})\bigr)\le I(\rho_{AB}).$$
